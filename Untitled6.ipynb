```python
import pandas as pd
import numpy as np
df=pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
print(df.head())
print(df.shape)
```

```python
print(df.head())
print(df.shape)
```

```python
# How many customers left vs stayed?
print(df['Churn'].value_counts())
```

```python
print(df['Churn'].value_counts(normalize=True)*100)
```

```python
#checking null values
print(df.isnull().sum())
```

```python
print(df.groupby('gender')['Churn'].value_counts())
```

```python
print(df.groupby('gender')['Churn'].value_counts())
```

```python
print(df.groupby('Contract')['Churn'].value_counts())
```

```python
print(df.groupby('Churn')['MonthlyCharges'].mean())
```

```python
print(df.groupby('Churn')['tenure'].mean())
```

```python
print(df.groupby('SeniorCitizen')['Churn'].value_counts())
```

```python
import numpy as np
print("Average monthly charges using numpy")
print(np.mean(df['MonthlyCharges']))
print("\nHow much charges vary between customers:")
print(np.std(df["MonthlyCharges"]))
print("\nMinimum Monthly Charge:")
print(np.min(df["MonthlyCharges"]))
print("\nMaximum Monthly Charges")
print(np.max(df["MonthlyCharges"]))



```
